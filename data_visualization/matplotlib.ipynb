# Object Oriented way

Matplotlib allows the aspect ratio, 
DPI (depth per inch) and figure size to be specified when the Figure object is created. 
You can use the figsize and dpi keyword arguments.
figsize is a tuple of the width and height of
the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1]) #([x,y,width%,hight%])
axis1.plot(x,y)
axis2 = fig.add_axes([0.4,0.6,.3,.3])#([40% right, 60% up, 30% width, 30% hight])
axis2.plot(x,y)
```

```python
fig = plt.figure(figsize=(6,6), dpi=100)
axis1 = fig.add_axes([0.1,0.1,0.6,0.8]) #([x,y,width%,hight%])
axis1.plot(x,y)
axis2 = fig.add_axes([0.5,0.5,0.3,0.3])#([40% right, 60% up, 30% width, 30% hight])
axis2.plot(x,y)
```

```python
# figsize means, overall canvas size, dpi means pixel add krna for more quality
```

# working with image

```python
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('images.jpg')
#img --  it will divide image in basis of pixel, which is in array 
plt.imshow(img)
plt.axis('off')
```

```python
cropped_img = img[500:800, 300:500]  # crop: rows 500-800, cols 300-500
plt.imshow(cropped_img)     #Row slicing: rows 500 to 799 (not 800)
plt.axis('off')         #Column slicing: columns 300 to 499 (not 500)
```
