```python
# how many countries_long have republic in full name
```

```python
count=0
def counting(txt):
    global count
    if 'republic' in txt.lower():
        count+=1
    return txt
df['country_long']=df['country_long'].apply(counting)
print(count)
```

```python
# global count
# This allows the function to access and modify the global variable count,
# not a local copy.
# Without global, Python would create a new local count variable inside
# the function — and the global one wouldn’t change.
```
