## merging two dataframe 

```python
employees=pd.DataFrame({
    'employee_id':[1,2,3,4,5],
    'name':['john','anna','peter','linda','bob'],
    'department':['HR','IT','Finance','IT','HR']
})
# DATAFRAME 2 : SALARY INFORMATION
salaries = pd.DataFrame({
    'employee_id':[1,2,3,6,7],
    'salaries':[67000,70000,75000,86000,69000],
    'bonus':[5000,7000,10000,4000,6000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries, on='employee_id') #these can be done without on, but we know merge of two things have a common things, e.g = employeee_id
```

```python
pd.merge(employees,salaries, on='employee_id',how='inner') # this will give common column of two dataframe
```

```python
pd.merge(employees,salaries, on='employee_id',how='outer') # this will give whole row of dataframe by filling NaN 
```

```python
pd.merge(employees,salaries, on='employee_id',how='left') #left means, I want to merge on the basis of employee, it means it will include all id which contains in employees
```

```python
pd.merge(employees,salaries, on='employee_id',how='right') # right means, I want to merge on the basis of salaries, it means it will include all id which contains in salaries
```

## concatination of two dataframe

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2]) # it is in order [df2,df1], then print like this, if u don't mention axis then it will print in row wise
```

```python
pd.concat([df1,df2], axis=1) # now it will print in column wise
```

## joining of two dataframe

```python
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df1.join(df2,how='outer')
```

```python
df1.join(df2,how='inner')
```

```python
df1.join(df2,how='left')
```

```python
df1.join(df2,how='right')
```
